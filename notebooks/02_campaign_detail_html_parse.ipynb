## Loading all html files and parsing them

```python
data = []

# function for handling error in emplty page cases
def safe_text(tag, default=""):
    return tag.get_text(strip=True) if tag else default


for i in range(29):
    with open(f"C:/Users/ASUS/OneDrive/Desktop/Web Scraping/04_Campaigns-Scrape/data/page_number{i}.html", "r", encoding="utf-8") as f:
        html = f.read()
    soup = BeautifulSoup(html, 'html.parser')
    data.append({
        "campaign_name" : safe_text(soup.find("h5")),

        "target_amount" : safe_text(soup.select_one("div.raised-of-primary div.number-holder")),

        "amount_raised" : safe_text(soup.select_one("div.raised-figs")),

        "number_of_donors" : safe_text(soup.select_one("label#label-donors")),

        "emails_or_phone" : safe_text(soup.select_one("div.row.mid-donate.mt-3 > div"))        
        
    })
```

```python
df = pd.DataFrame(data)
```

```python
df["number_of_donors"].str.replace("Donors", "")
```

```python
import re
df["number_of_donors"].str.extract(r"(\d+)\s*donors", flags=re.IGNORECASE)
```

```python
df["number_of_donors"] = df["number_of_donors"].str.extract(r"(\d+)\s*donors", flags=re.IGNORECASE)
```

```python
df["amount_raised"] = df["amount_raised"].replace(r"[a-zA-Z]+", "", regex=True)
```
